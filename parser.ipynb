```python
from time import sleep
import requests
import pandas as pd
import os.path
from datetime import datetime
```

```python
if os.path.isfile('dataset.csv'):
    print('dataset.csv exists')
    dataset = pd.read_csv('dataset.csv')
else:
    print('dataset.csv does not exist')
    dataset = pd.DataFrame(columns={'symbol', 'price','time'})
```

```python
dataset
```

```python
key = "https://api.binance.com/api/v1/exchangeInfo"
  
# requesting data from url
data = requests.get(key)  
data = data.json()
print(data)
```

```python
currencies = []
for i in data['symbols']:
    if i['symbol'][4:]=='USDT':
        currencies.append(i['symbol'])
```

```python
price = "https://api.binance.com/api/v1/ticker/price?symbol="
time = "https://api.binance.com/api/v1/time"


for i in range(0, 10):
    for cur in currencies:
        data = requests.get(price+cur).json()
        symb, cur_price = data['symbol'], data['price']
        tau = requests.get(time).json()['serverTime']
        tau = datetime.fromtimestamp(tau/1000.0).strftime('%Y-%m-%d %H:%M:%S.%f')
        dataset = dataset.append({'symbol':symb, 'price':cur_price, 'time':tau},ignore_index=True)

        #print(f"{symb} price is {cur_price}, time is {tau}")
dataset['time'] = pd.to_datetime(dataset['time'],unit='ms')
```

```python
dataset.to_csv('dataset.csv',index=False)
```
