```python
from time import sleep
import requests
import pandas as pd
import os.path
from datetime import datetime
```

```python
if os.path.isfile('dataset_klines.csv'):
    print('dataset_klines.csv exists')
    dataset = pd.read_csv('dataset_klines.csv')
else:
    print('dataset.csv does not exist')
    dataset = pd.DataFrame(columns=['time', 'open','high','low','close'])
```

```python
key = "https://api.binance.com/api/v1/exchangeInfo"
  
# requesting data from url
data = requests.get(key)  
data = data.json()
print(data)
```

```python
currencies = []
for i in data['symbols']:
    if i['symbol'][4:]=='USDT':
        currencies.append(i['symbol'])
```

```python
url = "https://api.binance.com/api/v1/klines"
symbol = "ETHUSDT"
interval = "1m"
res = requests.get(url+'?symbol='+symbol+'&interval='+interval)
if res.status_code == 200:
    data = pd.DataFrame(res.json())
    m = pd.DataFrame()
    m['time'] = pd.to_datetime(data[0],unit = 'ms')
    m['open'] = data[1]
    m['high'] = data[2]
    m['low'] = data[3]
    m['close'] = data[4]
m.set_index('time')
```

```python
dataset = pd.concat([dataset, m])
dataset.set_index('time',inplace=True)
dataset
```

```python
dataset.to_csv('dataset_klines.csv')
```
